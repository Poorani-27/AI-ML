```python
import pandas as pd
```

```python
dataset = pd.read_csv('https://github.com/Poorani-27/AI-ML/raw/main/Dataset/Admission%20Chance.csv')
dataset
```

```python
dataset.columns
```

```python
y = dataset['Chance of Admit ']
x = dataset.drop(['Serial No', 'Chance of Admit '], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
model.fit(x,y)
```

```python
x_test
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error,  mean_squared_error
```

```python
mae=mean_absolute_error(y_test,y_predict)
print("Mean Absolute Error : ",mae)
mae=mean_absolute_percentage_error(y_test,y_predict)
print("Mean Absolute percentage Error : ",mae)
mae=mean_squared_error(y_test,y_predict)
print("Mean squared Error : ",mae)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_predict)
print("R-squared Score: ", r2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_predict, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Plotting the diagonal line
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted')
plt.show()
```
